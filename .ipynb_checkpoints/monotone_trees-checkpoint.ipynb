### Ensembles dominants générés par un attribut/label

```python
def dominant_sets_attribute(a_j, labeled_set):    
    '''
        a_j : index of attribute in labeled_set
        labeled_set : labeled set
        for each object w in labeled_set, return its dominant set generated by a_j  
    '''
    dominant_sets = dict()
    for i in range(0, labeled_set.size()):
        dominant_sets[i] = set()
        for k in range(0, labeled_set.size()):
            if labeled_set.getX(k)[a_j] >= labeled_set.getX(i)[a_j]:
                dominant_sets[i].add(k)
                
    return dominant_sets

def dominant_sets_label(labeled_set):
    '''
        labeled_set : labeled set
        for each object w in labeled_set, return its dominant set generated by its label  
    '''
    dominant_sets = dict()
    for i in range(0, labeled_set.size()):
        dominant_sets[i] = set()
        for k in range(0, labeled_set.size()):
            if labeled_set.getY(k) >= labeled_set.getY(i):
                dominant_sets[i].add(k)
                
    return dominant_sets
    
dsa_1 = dominant_sets_attribute(0, training_set)
dsa_2 = dominant_sets_attribute(1, training_set)
dsl = dominant_sets_label(training_set)
```

### Construction hiérarchique de mesures de discrimination

#### f-layer :

```python
class F_layer:
    '''
        object-wise local monotonicity measure 
    '''
    def __init__(self, a_j, labeled_set):
        '''
            a_j : index of attribute in labeled_set
            labeled_set : labeled set
            dsa : dictionnary containing, for each w_i in labeled_set, 
                    its dominant set generated by a_j
            dsa : dictionnary containing, for each w_i in labeled_set, 
                    its dominant set generated by its label
        '''
        self.a_j = a_j
        self.labeled_set = labeled_set
        self.dsa = self.dominant_sets_attribute()
        self.dsl = self.dominant_sets_label()
        
    def value(self, w_i):
        raise NotImplementedError
        
    def dominant_sets_attribute(self):    
        '''
            for each object w in labeled_set, return its dominant set generated by a_j  
        '''
        n = self.labeled_set.size()
        
        dominant_sets = dict()
        for i in range(0, n):
            dominant_sets[i] = set()
            for k in range(0, n):
                if self.labeled_set.getX(k)[self.a_j] >= self.labeled_set.getX(i)[self.a_j]:
                    dominant_sets[i].add(k)

        return dominant_sets

    def dominant_sets_label(self):
        '''
            for each object w in labeled_set, return its dominant set generated by its label  
        '''
        n = self.labeled_set.size()
        dominant_sets = dict()
        for i in range(0, n):
            dominant_sets[i] = set()
            for k in range(0, n):
                if self.labeled_set.getY(k) >= self.labeled_set.getY(i):
                    dominant_sets[i].add(k)

        return dominant_sets
```

$dsr(\omega_i)$ :

```python
class Dsr(F_layer):
    def __init__(self, a_j, labeled_set):
        super().__init__(a_j, labeled_set)
    
    def value(self, w_i):
        '''
            w_i : index of object in labeled_set
            return dsr value of w_i, a_j
        '''
        intersection = self.dsa[w_i].intersection(self.dsl[w_i])

        return len(intersection) * 1.0 / len(self.dsa[w_i])
```

$mindsr(\omega_i)$

```python
class Mindsr(F_layer):
    def __init__(self, a_j, labeled_set):
        super().__init__(a_j, labeled_set)
    
    def value(self, w_i):
        '''
            w_i : index of object in labeled_set
            return mindsr value of w_i, a_j
        '''
        n = self.labeled_set.size()
        intersections = []
        dominant_set = self.dsa[w_i]
        
        for w_h in dominant_set:
            intersections.append(len(self.dsa[w_h].intersection(self.dsl[w_h])))
    
    
        return min(intersections) * 1.0 / len(dominant_set)    
```

$maxdsr(\omega_i)$

```python
class Maxdsr(F_layer):
    def __init__(self, a_j, labeled_set):
        super().__init__(a_j, labeled_set)
    
    def value(self, w_i):
        '''
            w_i : index of object in labeled_set
            return maxdsr value of w_i, a_j
        '''
        n = self.labeled_set.size()
        intersections = []
        dominant_set = self.dsa[w_i]
        
        for w_h in dominant_set:
            intersections.append(len(self.dsa[w_h].intersection(self.dsl[w_h])))
    
    
        return max(intersections) * 1.0 / len(dominant_set)    
```

$avgdsr(\omega_i)$

```python
class Avgdsr(F_layer):
    def __init__(self, a_j, labeled_set):
        super().__init__(a_j, labeled_set)
    
    def value(self, w_i):
        '''
            w_i : index of object in labeled_set
            return avgdsr value of w_i, a_j
        '''
        n = self.labeled_set.size()
        intersections = []
        dominant_set = self.dsa[w_i]
        
        for w_h in dominant_set:
            intersections.append(len(self.dsa[w_h].intersection(self.dsl[w_h])))
            
        return (1.0/len(dominant_set) * np.sum(intersections)) * 1.0 / len(dominant_set)    
```

#### g-layer :

```python
class G_layer:
    '''
        object-wise local non-monotonicity measure
    '''
    def __init__(self, labeled_set):
        '''
            labeled_set : labeled set
        '''
        self.labeled_set = labeled_set
        
    def value(self, f_value):
        raise NotImplementedError
```

```python
class Log(G_layer):
    def __init__(self, labeled_set):
        super().__init__(labeled_set)
    
    def value(self, f_value):
        '''
            f_value : value computed by f_layer
            return -log_2(f_value)
        '''
        return -log(f_value, 2) 
```

```python
class One_minus(G_layer):
    def __init__(self, labeled_set):
        super().__init__(labeled_set)
    
    def value(self, f_value):
        '''
            f_value : value computed by f_layer
            return 1 - f_value
        '''
        return 1 - f_value
```

#### h-layer :

```python
class H_layer:
    '''
        aggregated local non-monotonicity measure
    '''
    def __init__(self, labeled_set):
        '''
            labeled_set : labeled set
        '''
        self.labeled_set = labeled_set
        
    def value(self, g_values):
        raise NotImplementedError
```

```python
class Sum(H_layer):
    def __init__(self, labeled_set):
        super().__init__(labeled_set)
    
    def value(self, g_values):
        '''
            return (1/labeled_set.size()) * sum(g_values)
        '''
        return (1.0/self.labeled_set.size()) * np.sum(g_values)
```

#### Rank Shannon discrimination measure $H^*_s(\lambda | a_j)$ :

```python
def rsdm(f_layer, labeled_set):
    '''
        f_layer : object-wise local monotonicity measure 
        labeled_set : labeled set
        return the rank shannon discrimination measure generated by attribute a_j
    '''
    H = 0
    n = labeled_set.size()
    for i in range(0, n):
        H += -log(f_layer.value(i), 2)
    return (1.0/n)*H
```

```python
dsr_shannon_1 = dsr(0, training_set)
dsr_shannon_2 = dsr(1, training_set)
print("Rank Shannon discrimination measure generated by first attribute : ", rsdm(dsr_shannon_1, training_set))
print("Rank shannon discrimination measure generated by second attribute : ", rsdm(dsr_shannon_2, training_set))
```

#### Rank Gini discrimination measure $H^*_G(\lambda | a_j)$

```python
def rgdm(f_layer, labeled_set):
    '''
        f_layer : object-wise local monotonicity measure 
        labeled_set : labeled set
        return the rank gini discrimination measure generated by attribute a_j
    '''
    H = 0
    n = labeled_set.size()
    for i in range(0, n):
        H += 1-(f_layer.value(i))
    return (1.0/n)*H
```

```python
dsr_gini_1 = dsr(0, training_set)
dsr_gini_2 = dsr(1, training_set)
print("Rank Gini discrimination measure generated by first attribute : ", rgdm(dsr_gini_1, training_set))
print("Rank Gini discrimination measure generated by second attribute : ", rgdm(dsr_gini_2, training_set))
```

#### Generic rank discrimination measure :

```python
class Rdm:
    '''
        Generic rank discrimination measure
    '''
    def __init__(self, h, g, f):
        '''
            h : object-wise local monotonicity measure 
            g : object-wise local non-monotonicity measure 
            f : aggregated local non-monotonicity measure
            labeled_set : labeled set
        '''
        self.h = h 
        self.g = g
        self.f = f
    
    def value(self, labeled_set):
        g_f = []
        
        for i in range(0, labeled_set.size()):
            g_f.append(self.g.value(self.f.value(i)))
        
        return self.h.value(g_f)
```

```python
f_layer = Dsr(1, training_set)
g_layer_shannon = Log(training_set)
g_layer_gini = One_minus(training_set) 
h_layer = Sum(training_set)

rsdm = Rdm(h_layer, g_layer_shannon, f_layer)
rgdm = Rdm(h_layer, g_layer_gini, f_layer)

print("Rank Shannon discrimination measure generated by second attribute : " , rsdm.value(training_set))
print("Rank Shannon discrimination measure generated by second attribute : " , rgdm.value(training_set))
```

```python
def discretize(rdm, labeled_set):
    '''
        rdm : rank discrimination measure
    '''
    a_j = rdm.f.a_j
    min_entropy = 1.1
    min_threshold = 0.0    
    n = labeled_set.size()
    ind = np.argsort(labeled_set.x,axis=0)[:,a_j] # sort values 
    print(ind)
    binary_set = LabeledSet(labeled_set.getInputDimension())
    binary_set.x = labeled_set.x.copy()
    binary_set.x[:,a_j] = np.ones(labeled_set.size())
    binary_set.y = labeled_set.y
    
    for i in range(n-1):
        current = labeled_set.getX(ind[i])[a_j]
        lookahead = labeled_set.getX(ind[i+1])[a_j]
        binary_set.x[ind[i]][a_j] = 0
        
        if current == lookahead:
            continue
        threshold_val = (current + lookahead) / 2.0
        
        H = rdm.value(binary_set)
        print(threshold_val, H)
        if H < min_entropy:
            min_entropy = H
            min_threshold = threshold_val
    return (min_threshold, min_entropy)
```
