```python
g = Log()
h = Sum()

# rank Shannon discrimination measure
f_r = Dsr()
rsdm = Gdm(h, g, f_r)

# conditional Shannon entropy
f = Ds()
sdm = Gdm(h, g, f)

# pessimistic rank discrimination measure
f_p = Mindsr()
g_p = Frac() 
prdm = Gdm(h, g_p, f_p)

# rank Gini discrimination measure 
g_g = One_minus()
rgdm = Gdm(h, g_g, f_r)

# Gini discrimination measure
gdm = Gdm(h, g_g, f)

# H_M
h_m = Max()
g_m = One_minus_square()
H_m = Gdm(h_m, g_m, f_r)

# H_Q
f_q = Avgdsr()
h_q = Square_root()
H_q = Gdm(h_q, g_g, f_q)
```

```python
dataset = generate_1Ddataset(6, 100)
```

```python
discretize(rsdm, dataset,0)
```

```python
h1 = []
h2 = []
h3 = []
h4 = []
h5 = []
h6 = []
h7 = []
thresholds = []

for i in range(100):
    dataset = generate_1Ddataset(3, 100)
    t, h = discretize_steps(rsdm, dataset, 0)
    thresholds.extend(t)
    h1.extend(h)
    h2.extend(discretize_steps(sdm, dataset, 0)[1])
    h3.extend(discretize_steps(rgdm, dataset, 0)[1])
    h4.extend(discretize_steps(gdm, dataset, 0)[1])
    h5.extend(discretize_steps(prdm, dataset, 0)[1])
    h6.extend(discretize_steps(H_m, dataset, 0)[1])
    h7.extend(discretize_steps(H_q, dataset, 0)[1])

h1 = [h for _,h in sorted(zip(thresholds,h1))]
h2 = [h for _,h in sorted(zip(thresholds,h2))]
h3 = [h for _,h in sorted(zip(thresholds,h3))]
h4 = [h for _,h in sorted(zip(thresholds,h4))]
h5 = [h for _,h in sorted(zip(thresholds,h5))]
h6 = [h for _,h in sorted(zip(thresholds,h6))]
h7 = [h for _,h in sorted(zip(thresholds,h7))]

h1 = np.array(h1)
h2 = np.array(h2)
h3 = np.array(h3)
h4 = np.array(h4)
h5 = np.array(h5)
h6 = np.array(h6)
h7 = np.array(h7)
```

```python
df = pd.DataFrame({"H*_S": h1, "H_S": h2, "H*_G": h3, "H_G": h4, "H_P": h5, "H_M": h6, "H_Q": h7})

sns.pairplot(df)
```
